```python
from pandas import DataFrame
from sklearn import preprocessing
from torch import tensor

from game_lists_site.models import Game, User
from game_lists_site.utils.utils import get_cbr_for_game, get_mbcf_for_game
```

```python
game = Game.get_by_id(878750)
```

```python
cbr_result = get_cbr_for_game(game, 36)
mbcf_result = get_mbcf_for_game(game, 36)
```

```python
def normalize_dict(dict_data: dict, coef: float = 1):
    values = list(dict_data.values())
    values = preprocessing.normalize([values])[0] * coef
    return {k: v for k, v in zip(dict_data, values)}

# def merge_dicts_v2(dicts: list):
#     result = {}
#     for d in dicts:
#         for v in d:
#             if v in result:
#                 result[v] *= d[v]
#             else:
#                 result[v] = d[v]
#     result = {key: value/3 for key, value in sorted(result.items(), key=lambda item: item[1], reverse=True) }
#     return result

def merge_dicts(dicts: list):
    result = {}
    for d in dicts:
        for v in d:
            if v in result:
                result[v] += d[v]
            else:
                result[v] = d[v]
    result = {key: value for key, value in sorted(result.items(), key=lambda item: item[1], reverse=True) }
    return result
```
